Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no compiencen por 28, ya que estamos con códigos postales de Madrid

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dnetro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

````r
airbnb[which(airbnb$CodigoPostal == '27013'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '280013'),'CodigoPostal'] <- '28013'
airbnb[which(airbnb$CodigoPostal == '28002\n28002'),'CodigoPostal'] <- '28002'
airbnb[which(airbnb$CodigoPostal == '2805'),'CodigoPostal'] <- '28005'
airbnb[which(airbnb$CodigoPostal == 'Madrid 28004'),'CodigoPostal'] <- '28004'
airbnb[which(airbnb$CodigoPostal == '28051\n28051'),'CodigoPostal'] <- '28051'
airbnb[which(airbnb$CodigoPostal == '20013'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '27004'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '-'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '28'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '2815'),'CodigoPostal'] <- '28015'
airbnb[which(airbnb$CodigoPostal == '25008'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '2015'),'CodigoPostal'] <- NA
airbnb[which(airbnb$CodigoPostal == '2804'),'CodigoPostal'] <- '28004'
````

Mostrar una dataframe con el nombre del barrio, el precio y el número de entradas.

Ayuda: Podeis crear un nuevo dataframe con las columnas "Barrio" y "Freq" que contenga el número de entradas en cada barrio y hacer un merge con el dataframe del punto anterior.

````r
apartamentos_barrios <- data.frame(table(airbnb_entire$Barrio))
names(apartamentos_barrios)<- c("Barrio", "Freq")
apartamentos_barrios_2 <- merge(apartamentos_barrios, mean_price, by="Barrio")
(apartamentos_barrios_2)
````

Partiendo del dataframe anterior, muestra los 5 barrios con mayor precio, pero que tengan más de 100 entradas de alquiler.

````r
barrios_100 <- apartamentos_barrios_2[apartamentos_barrios_2$Freq > 100, ]
barrios_100 <- barrios_100[order(barrios_100$Precio, decreasing=TRUE, na.last=FALSE),]
barrios_100_5 <-head(barrios_100, 5)
barrios_100_5
````

Dibuja el diagrama de densidad de distribución de los diferentes precios. Serían 5 gráficas, una por cada barrio.

````r
barrios_100_5_vector <- c(barrios_100_5$Barrio)
barrios_100_5_vector <- droplevels(barrios_100_5_vector)
barrios_100_5_density <- airbnb_entire[airbnb_entire$Barrio %in% barrios_100_5_vector, ]

ggplot(data=barrios_100_5_density ,aes(x=Precio, , color=Barrio))+geom_density()
````

Calcula el tamaño medio, en metros cuadrados, para los 5 barrios anteriores y muestralo en el mismo dataframe junto con el precio y número de entradas

````r
barrios_100_5_metros <- aggregate(MetrosCuadrado ~ Barrio, barrios_100_5_density, mean, na.rm=TRUE)
apartamentos_barrios3 <- data.frame(table(airbnb_entire$Barrio))
names(apartamentos_barrios3)<- c("Barrio", "Freq")
barrios_100_5_metros_2 <- merge(barrios_100_5_metros, apartamentos_barrios3, by="Barrio")
barrios_100_5_metros_2 <- merge(barrios_100_5_metros, barrios_100_5, by="Barrio")
barrios_100_5_metros_2
````

Dibuja el diagrama de densidad de distribución de los diferentes tamaños de apartamentos. Serían 5 gráficas, una por cada barrio.

````r

barrios_100_5_metros_2_vector <- c(barrios_100_5_metros_2$Barrio)
barrios_100_5_metros_2_vector <- droplevels(barrios_100_5_metros_2_vector)
barrios_100_5_metros_2_density <- airbnb_entire[airbnb_entire$Barrio %in% barrios_100_5_metros_2_vector, ]

ggplot(data=barrios_100_5_metros_2_density,aes(x=MetrosCuadrado, , color=Barrio))+geom_density(bw=5)
````

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

````r
"Primero el test de Shapiro para comprobar si siguen una distribucion normal o no"
barrio_Goya <- barrios_100_5_metros_2_density[barrios_100_5_metros_2_density$Barrio == 'Goya',]
barrio_Justicia <- barrios_100_5_metros_2_density[barrios_100_5_metros_2_density$Barrio == 'Justicia',]
barrio_Recoletos <- barrios_100_5_metros_2_density[barrios_100_5_metros_2_density$Barrio == 'Recoletos',]
barrio_Sol <- barrios_100_5_metros_2_density[barrios_100_5_metros_2_density$Barrio == 'Sol',]
barrio_Trafalgar <- barrios_100_5_metros_2_density[barrios_100_5_metros_2_density$Barrio == 'Trafalgar',]


shapiro.test(barrios_100_5_metros_2_density$MetrosCuadrado)
shapiro.test(barrio_Goya$MetrosCuadrado)
shapiro.test(barrio_Justicia$MetrosCuadrado)
shapiro.test(barrio_Recoletos$MetrosCuadrado)
shapiro.test(barrio_Sol$MetrosCuadrado)
shapiro.test(barrio_Trafalgar$MetrosCuadrado)
````

````r
"En general, a excepcion del barrio_Goya, dado el bajo p-valor, vemos que, en lineas generales,no siguen una distribución gaussiana, por lo que realizariamos el test de Kruskal"

kruskal.test(MetrosCuadrado ~ Barrio, data = barrios_100_5_metros_2_density)
````

Calcular un modelo lineal que combine alguna de estas variables:
* NumBanyos
* NumDormitorios
* MaxOcupantes
* MetrosCuadrados

````r
barrio_sol<-subset(airbnb_entire,Barrio=="Sol")

modelo_lineal <- lm(data=barrio_sol, formula = MaxOcupantes ~ MetrosCuadrado)
summary(modelo_lineal)
````

Primero calculamos la correlación para ver como se relacionan estas variables entre sí.

````r
library(dplyr)
barrio_sol_2 <- barrio_sol |> select (-CodigoPostal, -TipoPropiedad,- TipoAlquiler, -NumCamas, -TipoCama, -Comodidades, -PiesCuadrados, -Precio, -Puntuacion)
cor(barrio_sol_2[2:ncol(barrio_sol_2)],use="pairwise.complete.obs")
````

Se observa que la correlación entre el número de dormitorios y los metros cuadrados es sorprendentemente baja.
¿Son de fiar esos números?

Mediante un histograma o curvas de densidad podemos descartar números que notienen sentido en el dataframe barrio_sol,
para tener una matriz de correlación que tenga mayor sentido.


````r
ggplot(data=barrio_sol_2 ,aes(x=MetrosCuadrado))+geom_histogram(bins=20, color="blue")
````

````r
barrio_sol_3 <- barrio_sol_2[barrio_sol_2$MetrosCuadrado >15,]

ggplot(data=barrio_sol_3 ,aes(x=MetrosCuadrado))+geom_histogram(bins=20, color="blue")
````

````r
"Después de eliminar los alquileres de <15 m2, podemos ver que la correlación entre el número de dormitorios y los metros cuadrados es bastante más alta que en la anterior comprobación"
cor(barrio_sol_3[2:ncol(barrio_sol_2)],use="pairwise.complete.obs")
````

Una vez que hayamos filtrado los datos correspondientes calcular el valor o la combinación de valores que mejor nos permite obtener el precio de un inmueble.

````r
barrio_sol_4 <- barrio_sol |> select (-CodigoPostal, -TipoPropiedad,- TipoAlquiler, -NumCamas, -TipoCama, -Comodidades, -PiesCuadrados, -Puntuacion)
barrio_sol_4 <- barrio_sol_4[barrio_sol_4$MetrosCuadrado >15,]
cor(barrio_sol_4[2:ncol(barrio_sol_4)],use="pairwise.complete.obs")
````

¿Que variable es más fiable para conocer el precio de un inmueble, el número de habitaciones o los metros cuadrados?

````r
"De acuerdo al calculo de correlación del punto anterior, el nº de dormitorios, es la variable más fiable, para conocer el precio de un inmueble, vamos a ver con un modelo de regresión lineal:"
modelo<-lm(data=barrio_sol_4, formula = Precio ~ NumDormitorios+MetrosCuadrado)
summary(modelo)
"Si tenemos en cuenta el p-valor, la variable más fiable para obtener el precio de un inmbueble, y que se corresponde con la correlación calculada, es el nº de dormitorios"
````

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada metro cuadrado extra del piso?

````r
model_metros <- lm(data=barrio_sol_4, formula= Precio  ~ MetrosCuadrado)
summary(model_metros)
confint(model_metros)
"El precio del alquiler, aumenta entre 0.69 y 1.26 con una estimación de 0.98 (0.9796), por cada metro cuadrado extra del piso"
````

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada habitación?

````r
model_habitacion <- lm(data=barrio_sol_4, formula= Precio  ~ NumDormitorios)
summary(model_habitacion)
confint(model_habitacion)
"El precio del alquiler, aumenta entre 31.18 y 47.78 con una estimación de 39.48, por cada habitación"
````

¿Cual es la probabilidad de encontrar, en el barrio de Sol, un apartamento en alquiler con 3 dormitorios?
¿Cual es el intervalo de confianza de esa probabilidad?

````r
barrio_sol_r <- barrio_sol[,c("Barrio","NumDormitorios")]
barrio_sol_3dormitorios <- barrio_sol_r[barrio_sol_r$NumDormitorios == 3,]
barrio_ratio <- nrow(barrio_sol_3dormitorios)/nrow(barrio_sol_r)
barrio_ratio
"Las probabilidades de encontrar un alquiler con 3 dormitorios en Sol, son del 7,84%"
````

````r
binom.test(nrow(barrio_sol_3dormitorios),nrow(barrio_sol_r))
"El intervalo de confianza es entre el 5,96 y el 10,09 % de probabilidades de encontrar un alquiler con 3 dormitorios en Sol"
````
